```python
from utils import * 
import matplotlib.pyplot as plt
import time 
```

```python
from tensorflow.keras.datasets import cifar10
import numpy as np

# Load CIFAR-10 dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
X_train = X_train.astype(np.float32) / 255.0  # Normalize to [0,1]
X_test = X_test.astype(np.float32) / 255.0  # Normalize to [0,1]
```

```python
U, sigma, Vt = SVD_dataset(X_train)
```

```python
print(U.shape)
print(sigma.shape)
print(Vt.shape)
```

```python
plt.plot(sigma)
plt.yscale('log')
plt.title("CIFAR 10: Singular Value")
plt.grid(True)
```

```python
start_time = time.time()
reconstructed_image = U[20, :] @ np.diag(sigma) @ Vt
end_time = time.time()
print("Time taken to reconstruct image: ", end_time - start_time)
reconstructed_image = reconstructed_image.reshape(32, 32, 3)
plt.imshow(reconstructed_image)
plt.title("CIFAR 10: Reconstructed Image")
```
